# Exploratory Data Analysis

```python
data.columns
```

```python
fx=sns.countplot(x="Size",data=data,palette="Accent")
for bars in fx.containers:
    fx.bar_label(bars)
```

# Note - From above graph you can see that most of the people buy M-Size

```python
data.groupby(["Size"],as_index=False)["Quantity"].sum().sort_values(by="Quantity",ascending=False)
```

```python
qty=data.groupby(["Size"],as_index=False)["Quantity"].sum().sort_values(by="Quantity",ascending=False)
sns.barplot(x="Size",y="Quantity",data=qty)
```

```python
plt.figure(figsize=(9,5))
sns.countplot(x="Courier Status",data=data,hue="Status")
```

# Note - from above graph majority of the orders are shipped through the courier

```python


data["Size"].hist()
```

```python
data["Category"] = data["Category"].astype(str)
plt.figure(figsize=(10,5))
plt.hist(data["Category"],bins=20,edgecolor="black")
plt.show()
```

```python
b2b_check= data["B2B"].value_counts()
plt.pie(b2b_check,labels=b2b_check.index,autopct="%1.1f%%")
plt.show()
```

# note - from above chart we can see that maximum 99.2% of buyers are retailer and 0.8% are B2B buyer

```python
x = data["Category"]
y = data["Size"]

plt.scatter(x,y)
plt.xlabel("Category")
plt.ylabel("Size")
plt.title("scatter plot")
plt.show()
```

```python
plt.figure(figsize=(9,5))
sns.countplot(data=data,x="ship-state")
plt.title("Distribution of state")
plt.xticks(rotation=90)
plt.show()
```

```python
top_10_states= data["ship-state"].value_counts().head(10)

plt.figure(figsize=(9,5))
sns.countplot(data=data[data["ship-state"].isin(top_10_states.index)],x="ship-state")
plt.xticks(rotation=90)
plt.show()
```

# note - from above graph now you can see that most buyers are maharashtra state
